```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../input/data11/advertising.csv')
df.head(20)
```

```python
X = df.iloc[:,2:4]
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train) 
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear',random_state = 0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
acc                                           # ---------(1)
```

```python
# OR
classifier.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'kernel':['linear'],'C':[10,50,150,250,350,500,700,1000]},
              {'kernel':['rbf'],'gamma':[0.001,0.008,0.01,0.1,0.5,0.8,0.9],'C':[10,50,150,250,350,500,700,1000]}]
gscv = GridSearchCV(estimator = classifier,param_grid = parameters,scoring  = 'accuracy',cv = 10)
gscv.fit(X_train,y_train)
```

```python
print(gscv.best_params_,'\n',gscv.best_score_)
```

```python
# As we find that the best kernel is rbf instead of linear with gamma value of 0.1, so we replace classifier variable by--

classifier = SVC(kernel = 'rbf',gamma = 0.1,random_state = 0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
acc                                      # --------(2)
```

**Conclusion -  From (1) & (2) we find that the accuracy of our model increased to 0.92 from 0.85**
